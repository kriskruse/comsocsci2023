```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph() # creates the graph
G.add_node(1) # adds a single node
G.add_nodes_from([1, 2, 3, 4]) # adds nodes using it as an iterator
G.add_edge(1, 2) # adds a single edge
G.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4), (3, 4)]) # adds edges using it as an iterator

MG = nx.MultiGraph() # creates a multigraph, multigraphs can have multiple edges between nodes
MG.add_edge(1, 2, weight=0.5) # adds a single edge with weight, can also be done in a normal graph
MG.add_edges_from([(1, 2),(2, 1) ,(1, 3), (2, 3), (3, 2), (2, 4), (3, 4)], weight=0.5) # adds edges using it as an iterator

subax = plt.subplot(121) # creates a subplot
nx.draw(G,with_labels=True , ax=subax) # draws the graph
subax2 = plt.subplot(122)
nx.draw(MG,with_labels=True , ax=subax2) # draws the multigraph
```
